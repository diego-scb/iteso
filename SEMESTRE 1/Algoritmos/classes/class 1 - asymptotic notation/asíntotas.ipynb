
# Asymptotes

```python
import math
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0.1, 7., .1)
lg2 = [math.log2(x) for x in n]
nlg2 = [x*math.log2(x) for x in n]
labels=["logarithmic","linear","quasilinear","quadratic","cubic","exponential"]

def graficar():
    plt.figure(figsize=(10,6))
    plt.plot(n,lg2,'g-', label=labels[0] )
    plt.plot(n, n, 'b--', label=labels[1])
    plt.plot(n, nlg2, 'c-', label=labels[2])
    plt.plot(n, n**2, 'k-', label=labels[3])
    plt.plot(n, n**3, 'y-', label=labels[4])
    plt.plot(n, 2**n, 'r-', label=labels[5])
    plt.legend()
    plt.show()

graficar()

# incrementar el rango para comparar
n = np.arange(0.1, 15., .1)
lg2 = [math.log2(x) for x in n]
nlg2 = [x*math.log2(x) for x in n]

graficar()
```

###  Compare $n^{10}$ vs $2^n$

```python
t = np.arange(0.1, 60., .1)
plt.figure(figsize=(10,6))
# red dashes, blue squares and green triangles
plt.plot(t, t**10, 'm-', label="polynomial n^10")
plt.plot(t, 2**t, 'r-', label="exponential 2^n")
plt.legend()
```

if the input is doubled, how much does the execution time increase?

```python
t = np.arange(0.1, 40., .5)

plt.figure(figsize=(10,6))

plt.plot(t, t, 'b-', label="T(N)=linear")
plt.plot(t, 2*t, 'y-', label="2T(N)=T(2N)")
plt.legend()
```

if the input is doubled in quasilinear complexity

```python
t = np.arange(0.1, 40., .5)
lg2 =  [math.log2(x) for x in t]
nlg2_n = [x*math.log2(x) for x in t]
nlg2_2n = [2*x*math.log2(2*x) for x in t]

plt.figure(figsize=(10,6))
plt.plot(t, nlg2_n, 'b-', label="T(N)=quasilinear Nlg(N)")
plt.plot(t, 2*np.asarray(nlg2_n), 'y-', label="2T(N) double the time")
plt.plot(t, nlg2_2n, 'c-', label="T(2*N)=(2N)log_2(2N) double the input")
plt.plot(t, 3*np.asarray(nlg2_n), 'r-', label="3T(N) triple the time")
plt.legend()
```

```python
# LOGARITHMIC
#log(100) = 1h
#log(x)  = 2h

x=2**(2*math.log2(100))
x
```

```python
# LINEAR
# 100 = 1h
# x = 2h

x = 2*100
x
```

```python
# QUASILINEAR
# 100 log (100 ) = 1h 
# (x)log(x)     = 2h
t = np.arange(100, 200, 1)
nlgn= [x*math.log2(x) for x in t]
objetivo= 2*(100*math.log2(100))
plt.plot(t,nlgn,'g*')

n =100
val= n*math.log2(n)
while(val < objetivo):
    n=n+1
    val = n*math.log2(n)
print(n)
```

Excercice $ f(n)=12N + 9  \in O(n^2) $

```python
#f(n)=12N + 9  
t = np.arange(1, 20, 1)
plt.plot(t, 12*t+9, 'b-', label="$f(n)=12n+9$")
plt.plot(t, 10*t**2, 'r-', label="$10 (g(n)=2n^2) $")
plt.plot(t, 9*t**2, 'y-', label="$9g(n)=n^2$")
plt.legend()
```

Excercice $ f(n)=nlog_2n +5N  \in O(n^2) $

```python
#f(n)=Nlog_2 (n) + 5N  
t = np.arange(1, 20, 1)
lg2 = [x*math.log2(x)+5*x for x in t]
plt.plot(t, lg2, 'b-', label="$f(n)=nlog_2n +5N$")
#c1 = 3
plt.plot(t, 3*t**2, 'r-', label="$3g(n)=3n^2$")
plt.legend()
```

```python
#Theta(N^3)
#f(n)=2n^3+4N^2 
t = np.arange(1, 20, 1)
plt.plot(t, 2*t**3+4*t**2, 'b-', label="$f(n)=2n^3+4n^2$")
plt.plot(t, 2*t**3, 'r-', label="$c_0g(n)=2n^3$")
plt.plot(t, 3*t**3, 'g-', label="$c_1g(n)=3n^3$")
plt.legend()
```
