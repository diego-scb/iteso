# Exercise 1: Insertion Sort

```python
def insertionSort(arr, verbose=False):
    if verbose: print(arr)
    moves = comp = 0
    for i in range(1, len(arr)):
        temp = arr[i]
        moves += 1  # counting the movement from array to temporary
        j = i - 1
        #comp += 2  # for the sake of this task, we consider this only in the best case (when not entering the while loop)
        while j >= 0 and temp < arr[j]:
            comp += 2
            arr[j + 1] = arr[j]
            moves += 1  # counting the movement from array to temporary
            j -= 1
            if verbose: print(arr)
        
        arr[j + 1] = temp
        moves += 1  # counting the previous assignment
        if verbose: print(arr)
    return comp, moves

arr2 = [97, 53, 43, 32, 29, 16, 10]
arr1 = [10, 16, 29, 32, 43, 53, 97]
insertionSort(arr2, True)

```


## 1.A) A priori Analysis of Sorted Array

* ## Comparisons

    Here, the procedure is described as $\sum_{i=1}^{n-1} 2$ because in the first iteration of the while loop we do 2 comparisons : j >= 0 and temp < arr[j].
    $$\sum_{i=1}^{n-1} 2= 2 \times \left(\sum_{i=1}^{n-1} 1\right) = 2 \times (n-1) = 2n - 2$$
    
* ## Movements

    Here, the procedure is described as $\sum_{i=1}^{n-1} 2$ because we just do the move before entering the while loop and the one after the while loop.
    $$\sum_{i=1}^{n-1} 2 = 2 \times \left(sum_{i=1}^{n-1} 1 \right) = 2n - 2$$
    

## 1.B) A priori Analysis of Reversed Array

* ## Comparisons

    Here, the procedure is described as $\sum_{i=1}^{n-1}\left(\sum_{0}^{i-1}2 \right)$, because in each iteration of the wile loop we perform 2 comparisons.
    $$ \sum_{i=1}^{n-1}\left(\sum_{0}^{i-1}2 \right) = 2 \times \left(\sum_{i=1}^{n-1}\left(\sum_{0}^{i-1}1 \right)\right) = 2 \times \left(\sum_{i=1}^{n-1} i\right) = 2 \times \frac{(n-1) \times n }{2} = n^2 - n$$
    
* ## Movements

    Here, the procedure is described as $\sum_{i=1}^{n-1}\left(2 + \sum_{0}^{i-1}1 \right)$
    $$\sum_{i=1}^{n-1}\left(2 + \sum_{0}^{i-1}1 \right) = \sum_{i=1}^{n-1}\left(2 + i \right) = \sum_{i=1}^{n-1}2 + \sum_{i=1}^{n-1}i = 2n - 2 + \frac{(n-1) \times n}{2} = 2n - 2 + \frac{n^2}{2} - \frac{n}{2} = \frac{n^2}{2} + \frac{3n}{2} - 2$$
    



## 1.C) A Posteriori Analysis

The a posteriori analysis involves running the algorithm under test multiple times for various input array sizes. For each input size, multiple runs of the algorithm are performed using random arrays, and an average is obtained for both movements and comparisons. This process is repeated for different array sizes. The averages are plotted, revealing a trend.

```python
# Here the code is completed to gather all the data for plotting

# This function creates random arrays of size 'size' with values between minVal and maxVal
import random 
def createIntArray(size, minVal, maxVal):
    return [random.randint(minVal, maxVal) for x in range(size)]

#---------------------------------------------------------------------

''' minSize, maxSize: array sizes ranging from minSize to maxSize-1
    step: interval between array sizes: minSize, minSize+step, minSize+step*2, ..., maxSize
    runs: number of runs for each array size 
    Note: For each run, a random array is created and the sorting algorithm is executed'''
    
def dataForPlot(sortalgo, minSize=2, maxSize=100, step=1, runs=200):
    totalC = []# An array to store the average comparisons for each array size
    totalM = []# An array to store the average movements for each array size
    
    # Loop through different array sizes, tip: range(minSize, maxSize, step):
    
    for size in range(minSize, maxSize, step):
        
        #initialize the sum of comparison and moves variables 
        
        sumC = 0
        sumM = 0
                
        # Run the algorithm multiple times for each array size (in the range of runs )
        for i in range(runs):
        
            # Create a random array of the given size
            randarr = createIntArray(size, 0, 5*maxSize)
            # Run the sorting algorithm and get comparisons and movements
            comp, moves = sortalgo(randarr)
            #sum the comparisos
            sumC = sumC + comp
            #sum the moves
            sumM = sumM + moves
        
        # Calculate the average comparisons and movements
        
        avC = sumC / runs
        avM = sumM / runs
             
        # Store the average values in the respective arrays totalC and totalM
        
        totalC.append(avC)
        totalM.append(avM)
               
    # Return the arrays containing the average comparisons and movements
    return totalC, totalM

import numpy as np
import matplotlib.pyplot as plt

minSize = 5
maxSize = 400
step = 5

totalC, totalM = dataForPlot(insertionSort, minSize, maxSize, step, 20)
sizes = np.arange(minSize, maxSize, step)



plt.plot(sizes, totalC, "*", label="comparisons")
plt.plot(sizes, totalM, "*", label="moves")
plt.plot(sizes, [s**2 for s in sizes], label="quadratic")
plt.plot(sizes, [s*(np.log2(s)) for s in sizes], label="quasilinear")
plt.plot(sizes, [(s**2)/2 for s in sizes], label="1/2 quadratic")
plt.plot(sizes, [(s**2)/4 for s in sizes], label="1/4 quadratic")
plt.legend()
plt.show()

```

# Exercise 2: Bubble Sort

```python
def bubbleSort(arr, verbose=False):
    if verbose: print(arr)
    moves = comp = 0
    swapped = True
    for i in range(len(arr)-1):
        # Here is a comparison, but we won't consider it to make the task easier for you.
        if swapped: 
            swapped = False
            for j in range(len(arr)-i-1):
                comp += 1
                if arr[j] > arr[j+1]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
                    moves += 3  # considering 3 movements
                    swapped = True
                    if verbose: print(arr)
        else:
            break
            
    return comp, moves

arr2 = [97, 53, 43, 32, 29, 16]
arr1 = [16, 29, 32, 43, 53, 97]
bubbleSort(arr2, True)
```


## 2.A) A priori Analysis of Sorted Array

* ## Comparisons

    Here, the procedure is described as $\sum_{j=0}^{n-2}1$ because after the first iteration of the first for loop, the value of swapped is False. Therefore we never go into the first if statement in the rest of the iterations of the first for loop. So we just perform the conparisons from the second if statement in each iteration of the second for loop when i = 0.
    $$\sum_{j=0}^{n-2}= n - 1$$
    
* ## Movements

    Here, the procedure is described as $0$ because we never go into the second if statement where the moves are made.
    
    

## 2.B) A priori Analysis of Reversed Array

* ## Comparisons

    Here, the procedure is described as $\sum_{i=0}^{n-2} \left(\sum_{j=0}^{n-2-i}1 \right)$
    $$\sum_{i=0}^{n-2} \left(\sum_{j=0}^{n-2-i}1 \right) = \sum_{i=0}^{n-1} (n-1-i) = (n-1) \times \sum_{i=0}^{n-2}1 - \sum_{i=0}^{n-2} i = (n-1)^2 - \frac{(n-2)\times (n-1)}{2} = (n-1) \times \left((n-1) - \frac{(n-2)}{2}\right) = (n-1) \times \frac{n}{2} = \frac{n^2}{2} - \frac{n}{2}$$
    
* ## Movements

    Here, the procedure is described as $\sum_{i=0}^{n-2} \left(\sum_{j=0}^{n-2-i}3 \right)$
    $$\sum_{i=0}^{n-2} \left(\sum_{j=0}^{n-2-i}3 \right) = 3 \times \left(\sum_{i=0}^{n-2} \left(\sum_{j=0}^{n-2-i}1 \right) \right) = 3\times (\frac{n^2}{2} - \frac{n}{2}) = \frac{3n^2}{2} - \frac{3n}{2}$$

# 3. Selection sort

```python
def selectionSort(arr, verbose=False):
    if verbose:
        print(arr)
    moves = comp = 0
    for i in range (len(arr)-1):
        minV = i
        for j in range(i+1,len(arr)):
            comp+=1
            if(arr[j]<arr[minV]):
                minV = j
            
        if(minV != i):
            moves+=1
            arr[i], arr[minV] = arr[minV],arr[i]
            if verbose:
                print(arr)
    return comp,moves
            
selectionSort([97,53,43,32,29,16])
```

Perform an **a posteriori** analysis of the shell sort algorithm for comparisons and movements.

- Implement the shellsort (you can base it on code you find online)
- With array sizes: 100 to 4000 with intervals of 10.
- Perform 100 runs for each array size.
- Plot and display the polynomial.
- *If you obtain a score greater than 100, you will be capped at 100.*

```python
def shellSort(arr, verbose=False):
    
    n = len(arr)
    
    comp = 0
    moves = 0
    
    if verbose:
        print(arr)
    
    interval = (n+1) // 2
    while interval > 0:
        comp = comp + 1
        for i in range(interval, n):
            temp = arr[i]
            moves = moves + 1
            j = i
            while j >= interval and arr[j - interval] > temp:
                comp = comp + 2
                arr[j] = arr[j - interval]
                moves = moves + 1
                if verbose:
                    print(arr)
                
                j -= interval

            arr[j] = temp
            moves = moves + 1
            if verbose:
                print(arr)
            
        interval //= 2
    if verbose:
        print(arr)
    return comp, moves

arr2 = [97, 53, 43, 32, 29, 16, 10, 5]
arr1 = [5, 10, 16, 29, 32, 43, 53, 97]
shellSort(arr2, True)
```
