```python
def intersection(arr1, arr2):
    if len(arr1) < len(arr2):
        smallarr = arr1
        bigarr = arr2
    else:
        smallarr = arr2
        bigarr = arr1
    
    inter = []
    last_element = "not yet initialized"
    bigindex = 0
    for element in smallarr:
        if element == last_element:
            continue
        else:
            last_element = element
        while bigindex < len(bigarr)-1 and element != bigarr[bigindex]:
            bigindex = bigindex + 1
        if bigindex == len(bigarr)-1:
            bigindex = 0
        elif element == bigarr[bigindex]:
            inter.append(element)
    return inter


lengths = input()
arr1 = input()
arr2 = input()
arr1 = list(map(int,arr1.split()))
arr2 = list(map(int,arr2.split()))

output = intersection(arr1, arr2)

print(*output)
```
